```python
import requests
import pandas as pd
import sqlalchemy
```

```python
organic_dairy_cows_eu_prel = pd.read_csv('organic_dairy_cows_eu.csv', header=[1], skiprows=[0], delimiter=';')
```

```python
organic_dairy_cows_eu_prel.head(15)
```

```python
# Drop the first nine rows and last six rows
organic_dairy_cows_eu_prel = organic_dairy_cows_eu_prel.iloc[10:-6]
```

```python
# Reset the index
organic_dairy_cows_eu_prel.reset_index(drop=True, inplace=True)
```

```python
# Replace 'e' and 'p' values with NaN
organic_dairy_cows_eu_prel.replace(['e', 'p'], pd.NA, inplace=True)
```

```python
# Drop columns with all NaN values
organic_dairy_cows_eu_prel = organic_dairy_cows_eu_prel.dropna(axis=1, how='all')
```

```python
organic_dairy_cows_eu_prel.head(5)
```

```python
# Set the appropriate column names
columns = ['Country'] + [str(year) for year in range(2012, 2023)]
organic_dairy_cows_eu_prel.columns = columns
```

```python
organic_dairy_cows_eu_prel
```

```python
# List of EU 27 countries
eu_27_countries = [
    'Belgium', 'Bulgaria', 'Czechia', 'Denmark', 'Germany', 'Estonia', 'Ireland', 
    'Greece', 'Spain', 'France', 'Croatia', 'Italy', 'Cyprus', 'Latvia', 'Lithuania', 
    'Luxembourg', 'Hungary', 'Malta', 'Netherlands', 'Austria', 'Poland', 'Portugal', 
    'Romania', 'Slovenia', 'Slovakia', 'Finland', 'Sweden'
]
```

```python
# Filter the DataFrame to keep only EU 27 countries
organic_dairy_cows_eu_prel = organic_dairy_cows_eu_prel[organic_dairy_cows_eu_prel['Country'].isin(eu_27_countries)]
```

```python
# Reset the index
organic_dairy_cows_eu_prel.reset_index(drop=True, inplace=True)
```

```python
organic_dairy_cows_eu_prel
```

```python
# Filter the DataFrame to exclude rows with zero values for Malta and Slovenia
organic_dairy_cows_eu_prel = organic_dairy_cows_eu_prel[
    (organic_dairy_cows_eu_prel['Country'] != 'Malta') & 
    (organic_dairy_cows_eu_prel['Country'] != 'Slovenia')
]
```

```python
# Reset the index
organic_dairy_cows_eu_prel.reset_index(drop=True, inplace=True)
```
