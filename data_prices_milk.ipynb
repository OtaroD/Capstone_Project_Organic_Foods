```python
# Import requests package:
import requests
```

```python
# Set URL as url:
url = 'https://www.ec.europa.eu/agrifood/api/rawMilk/prices?'
```

```python
# Create parameter dictionary:
parameters = {
    'beginDate': '01/01/2021',
    'endDate': '31/12/2023'}
```

```python
# Send get request and save in variable r:
r = requests.get(url, parameters)

# Print result:
print(r)
```

```python
# Print headers:
r.headers
```

```python
# Print cookies:
r.cookies
```

```python
# Print encoding:
r.encoding
```

```python
# Apply JSON decoder and save in prices_milk_prelim:
prices_milk_prelim = r.json()

# Print prices_milk_prelim:
prices_milk_prelim
```

```python
# Print type of prices_milk_prelim:
type(prices_milk_prelim)
```

```python
# Import pandas package:
import pandas as pd

# Transform list to dataframe:
prices_milk = pd.DataFrame(prices_milk_prelim)
```

```python
# We now want to upload the dataframe as a database to the server.

# Import sql_functions.py because we need some functions from that module:
import sql_functions as sqlf

# We need to restart the kernel and rerun at this point if we changed the module since we first imported it.
```

```python
# Create a variable called engine using the get_engine function:
engine = sqlf.get_engine()
```

```python
# We set the schema to our course name:
schema = 'hh_analytics_23_4'

# We set the table_name variable to our group name + the name of the dataframe:
table_name = 'organic_food_prices_milk'
```

```python
# We need psycopg2 for raising possible error message:
import psycopg2
```

```python
# Write records stored in the dataframe to SQL database:
if engine!=None:
    try:
        prices_milk.to_sql(name=table_name, # name of SQL table variable
                        con=engine, # engine or connection
                        schema=schema, # our class schema variable
                        if_exists='replace', # Drop the table before inserting new values
                        index=False, # Write DataFrame index as a column
                        chunksize=5000, # Specify the number of rows in each batch to be written at a time
                        method='multi') # Pass multiple values in a single INSERT clause
        print(f"The {table_name} table was imported successfully.")
    # Error handling
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
        engine = None
else:
    print('No engine')
```
