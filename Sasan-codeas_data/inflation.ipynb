```python
# show inflation rate for chosen countries
chosen_countries = ['Germany', 'Luxembourg','Finland', 'Moldova', 'Kosovo', 'Montenegro']


plt.figure(figsize=(12, 8))

for country in chosen_countries:
    plt.plot(df.columns, df.loc[country], linewidth=3, label=country)


plt.xlabel('Year')
plt.ylabel('Inflation Rate')
plt.title('Inflation Rate Over Years for Chosen Countries')
plt.legend()



plt.xticks(rotation=45)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# show the mean of inflation for the whole europe across years
mean_values = df.mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(mean_values.index, mean_values.values, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Mean Value')
plt.title('Mean Values Across Years(All european countires)')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
